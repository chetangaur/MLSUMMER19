```
pic1=cv2.imread('dd.jpeg')
```

```
scale=30                                        # for compression give value less than 100 and for decompression give more than 100 
```

```
width=int(pic1.shape[1]*scale/100)
```

```
height=int(pic1.shape[0]*scale/100)
```

```
dim=(width,height)
```

```
Cpic1=cv2.resize(pic1,dim)
```
