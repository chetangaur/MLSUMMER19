```
import pandas as pd
```

```
read_train=pd.read_csv("train_data.csv")
```

```
read_test=pd.read_csv("test_data.csv")
```

```
read_train.head(30)
```

```
read_test.head(30)
```

```
features=read_train.iloc[0:30,3:]
```

```
features
```

```
label=read_test.iloc[0:30,2]
label
```

```
from sklearn.tree import DecisionTreeClassifier
```

```
dtc=DecisionTreeClassifier()
```

```
trained=dtc.fit(features,label)
```

```
from sklearn.metrics import accuracy_score
```

```
test_data=read_test.iloc[0:30,3:]
```

```
prediction=trained.predict(test_data)
```

```
prediction
```

```
survival=accuracy_score(label,prediction)
```
